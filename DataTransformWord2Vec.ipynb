```python
import pandas as pd
from gensim.models import Word2Vec
from gensim.utils import simple_preprocess
import numpy as np
import pandas as pd

```

```python
# Load your dataset
df = pd.read_csv('Data\Imbalance\df_clean_labeled_tokenized.csv')

tokenized_reviews = df['tokenized_review'].apply(simple_preprocess)  # Convert to list of tokenized words

# Train Word2Vec model
word2vec_model = Word2Vec(sentences=tokenized_reviews, vector_size=100, window=5, min_count=1, workers=4)
# Parameters:
# - sentences: Tokenized reviews
# - vector_size: Dimensionality of the word vectors
# - window: Maximum distance between the current and predicted word within a sentence
# - min_count: Ignores words with a total frequency lower than this
# - workers: Number of CPU cores to use for training
def vectorize(sentence):
    words = sentence
    words_vecs = [word2vec_model.wv[word] for word in words if word in word2vec_model.wv]
    if len(words_vecs) == 0:
        return np.zeros(100)
    words_vecs = np.array(words_vecs)
    return words_vecs.mean(axis=0)

embedded_column = [vectorize(sentence) for sentence in tokenized_reviews]

df_embeddings = pd.DataFrame(embedded_column)
combined_df = pd.merge(df[['sentiment']], df_embeddings, left_index=True, right_index=True)
combined_df.to_csv('df_sentiment_embeddings.csv', index=False)
combined_df
```
