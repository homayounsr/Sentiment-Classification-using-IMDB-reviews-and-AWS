```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
import re
import numpy as np
import itertools
from reviewExtractor import *
import requests
```

```python
df = pd.read_table('data.tsv')
df = df.drop(columns=['ordering','title' , 'region' , 'language' , 'types', 'attributes' , 'isOriginalTitle'])
df = df.drop_duplicates()
df[50000:55000]
```

```python
# movie links
base_url = "https://www.imdb.com/title/"
#create url with unique title id
movie_links = [base_url + title_id + '/reviews?start=0' for title_id in df['titleId']]
print(len(movie_links[65000:68000]))
movie_soups = [getSoup(link) for link in movie_links[65000:68000]]
```

```python
# get a list of soup links if link is not empty []
movie_review_list = [getReviews(movie_soup) for movie_soup in movie_soups if getReviews(movie_soup)]
# get review text from the review link
review = [getReviewText(url) for url in movie_review_list if getReviewText(url)]
rate = [getRate(url) for url in movie_review_list if getReviewText(url)]
print("There are a total of " + str(len(review)) + " individual movie reviews")
```

```python
# construct a dataframe
review_df = pd.DataFrame({'review': review , 'rate':rate})
review_df
```

```python
# Append DataFrame to an existing CSV file
review_df.to_csv('UserReviewIMDB/userReviews.csv', index=False, mode='a', header=False)
print("Data Appended to CSV")
```
